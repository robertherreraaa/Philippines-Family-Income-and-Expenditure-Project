```python
# 2: Impact of Education on Income
# Do household heads with higher education levels earn more?
```

```python
# Calculate median income for each education level and get the top 10
top_10_education_levels = df.groupby('Household Head Highest Grade Completed')['Total Household Income'].median().sort_values(ascending=False).head(10).index

# Filter the DataFrame to include only the top 10 education levels
df_top_education = df[df['Household Head Highest Grade Completed'].isin(top_10_education_levels)]

plt.figure(figsize=(10, 8))
sns.boxplot(data=df_top_education, x='Total Household Income', y='Household Head Highest Grade Completed', order=top_10_education_levels)
plt.title('Box Plot of Total Household Income by Top 10 Education Levels')
plt.xlabel('Total Household Income')
plt.ylabel('Education Level')
plt.tight_layout()
plt.show()
```

```python
# Sir Josel
# I want to get the allocation for expenditures in Davao Region

df[['Region', 'Main Source of Water Supply']]
result_davao = df[['Region', ' Source of Water Supply']]
```

```python
result_davao
```

```python
# Filter data for "XI - Davao Region"
davao_df = df[df['Region'] == "XI - Davao Region"]

# Count the occurrences of each water supply type
davao_watersupply_counts = davao_df['Main Source of Water Supply'].value_counts()

# Create explode list for all wedges, setting a small value for each
explode_list = [0.05] * len(davao_watersupply_counts)

plt.figure(figsize=(12, 10)) # Increased figure size
plt.pie(davao_watersupply_counts, labels=davao_watersupply_counts.index, autopct='%1.1f%%', startangle=15, explode=explode_list, pctdistance=0.85)
plt.title('Allocation of Water Supply in XI - Davao Region')
plt.axis('equal')  # Ensures pie is a circle
plt.tight_layout()
plt.show()
```
